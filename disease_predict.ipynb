```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\pc\Downloads\Training.csv")
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
#Train test splt
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report,precision_score,f1_score,recall_score
```

```python
models={
    "decision":tree.DecisionTreeClassifier(),
    "Random forest":RandomForestClassifier(max_depth=2, random_state=42),
    "K Nearest Neighbor":KNeighborsClassifier(n_neighbors=3),
    "Support vector machine":SVC(gamma="auto"),
    "Logistic Regression":LogisticRegression(random_state=42),
}
```

```python
result=[]
for name, model in models.items():
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    cm=confusion_matrix(ytest,ypred)
    ac=accuracy_score(ytest,ypred)
    ps=precision_score(ytest,ypred,average='weighted')
    f1=f1_score(ytest,ypred,average='weighted')
    rc=recall_score(ytest,ypred,average='weighted')
    result.append({
        "model":name,
       # "confusion_matrix:":cm,
        "Accuracy score:":ac,
        "Precision score:":ps,
        "f1 Score:":f1,
        "recall score:":rc,
        
        
    })
```

```python
result
```

```python
ds=pd.DataFrame(result)
ds
```

```python
sorted_result=ds.sort_values(by='model')
sorted_result
```

```python
bestmodel=sorted_result.loc[0,"model"]
bestmodel=models[bestmodel]
bestmodel
```
